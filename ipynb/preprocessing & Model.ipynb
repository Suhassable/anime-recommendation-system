```python
import pandas as pd
import numpy as np
import re
```

```python
df=pd.read_csv(r"C:\Users\suhas\Downloads\Anime Planet.csv")
```

```python
df.head(2)
```

```python
#calculating percentage of null values in row
index=[]
for i in range (0,len(df)):
    per=df.iloc[i,].isna().sum()/df.shape[1]*100
    index.append(per)
    
```

```python
index
```

```python
ind=[]
for i,u in enumerate(index):
    if u>70:
        ind.append(i)
        
```

```python
#remove the null row 
df=df.drop(df.index[ind])
```

```python
df
```

```python
df=df.drop(['Unnamed: 0','Unnamed: 0.1','url'],axis=1)
```

```python
df
```

```python
# df = df.loc[:,['Gener','Demographic','Studio','Synopsis','Theme']]
```

```python
df.head()
```

```python
#replacing null value with unknown 
df['Gener'] = df['Gener'].replace(np.nan, 'unknown')
df['Demographic'] = df['Demographic'].replace(np.nan, 'unknown')
df['Synopsis'] = df['Synopsis'].replace(np.nan, 'unknown')
df['Studio'] = df['Studio'].replace(np.nan, 'unknown')
df['Theme'] = df['Theme'].replace(np.nan, 'unknown')
```

```python
#removing , in gener
genre=[]
for i in df['Gener']:
    genre.append(re.sub(',','', i))
```

```python
df['Gener']=genre
```

```python
#removing punctuation in synopsis
syn=[]
for i in df['Synopsis']:
    syn.append(re.sub(r'[,\.\n\r\'\"\[\]\,]','', i))
```

```python
df['Synopsis']=syn
```

```python
Theme=[]
for i in df['Theme']:
    Theme.append(re.sub(r'[,\.\n\r\'\"\[\]\,]','', i))
    
df['Theme']=Theme
```

```python
df
```

```python
#creating a combining col of selected col
df['Comb']=df['Gener']+' '+df['Demographic']+' '+df['Studio']+' '+df['Synopsis']+' '+df['Theme']
```

```python
df=df.reset_index()
```

```python
df
```

```python
#creating similarity matrix
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
vector=cv.fit_transform(df['Comb']).toarray()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
def create_similarity():
    data = df
    # creating a count matrix
    cv = CountVectorizer()
    count_matrix = cv.fit_transform(data['Comb'])
    # creating a similarity score matrix
    similarity = cosine_similarity(count_matrix)
    return data,similarity
```

```python
data,similarity=create_similarity()
```

```python
similarity
```

```python
def recommend(anime):
    index = df[df['Name'] == anime].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    names=[]
    for i in distances[1:6]:
        names.append(df.iloc[i[0]].Name)
    return names
```

```python
recommend("Gintama°")
```

```python
dict={}
```

```python
#creating a dictionary for every anime and its recommendation
for i in df["Name"]:
    dict[i]=recommend(i)
```
